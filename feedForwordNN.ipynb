```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Building a feed forword neural network
```

```python
# number of samples per class 
N_class = 500
# randomly initialize input for 3 classes
x1 = np.random.randn(N_class,2) + np.array([0, 2]) # first class is centerd at (0,2)
x2 = np.random.randn(N_class,2) + np.array([2, 2]) # second class is centerd at (2,2)
x3 = np.random.randn(N_class,2) + np.array([-2, 2]) # third class is centerd at (-2,2)
X =  np.vstack([x1, x2, x3]) # now X contains total 1500 samples 
#print(len(X))
#print(X)
#print(X.size)
```

```python
# randomly take pridicted label i.e. supervised learning
Y = np.array([0]*N_class + [1]*N_class + [2]*N_class)
#print(Y)
#print(len(Y))
#print(Y.size)
```

```python
# plot graph between both features
# The alpha blending value, between 0 (transparent) and 1 (opaque)
# c : color, sequence, or sequence of color, optional, default: ‘b’
# s = area, X[:,0] = first column, X[:,1] = second column
plt.scatter(X[:,0], X[:,1],  s=200,  c=Y,  alpha=.5)
plt.show()
```

```python
# randomly initialize weights
D = 2 # dimension of input
M = 3 # hidden layer size
K = 3 # number of classes
W1 = np.random.randn(D,M)
b = np.random.randn(M)
W2 = np.random.randn(M,K)
c = np.random.randn(K)
#print(W1)
#rint(b)
#print(W2)
#print(c)
```

```python
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
def tanh(a):
    expa = np.exp(a)
    return expa / expa.sum(axis=1, keepdims=True)
```

```python
def forword(X,W1,b,W2,c):
    z = (X.dot(W1) + b)
    A = z.dot(W2) + c
    Y = tanh(A)
    return Y
```

```python
# determine the classification rate
# num correct / num total
def classificationRate(Y, P):
    n_total = 0
    n_correct = 0
    for i in range(len(Y)):
        n_total +=1
        if Y[i] == P[i]:
            n_correct +=1
    return float(n_correct) / n_total        
```

```python
P_Y_given_X = forword(X,W1,b,W2,c)
print(P_Y_given_X)
```

```python
P = np.argmax(P_Y_given_X, axis=1)
#print(P)
```

```python
# verify we chose the correct axis
assert(len(P) == len(Y))
```

```python
print ("Classification rate for randomly chosen weights:", classificationRate(Y, P))
```
